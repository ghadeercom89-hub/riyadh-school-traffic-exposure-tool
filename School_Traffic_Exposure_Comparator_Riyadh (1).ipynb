# School Traffic Exposure Comparator – Riyadh
This notebook estimates inhaled PM2.5 dose for a child waiting near traffic outside a school.
It compares two scenarios (e.g., longer wait vs shorter wait).


```python
# Scenario A Inputs
background_A = float(input('Scenario A - Background PM2.5 (µg/m3): '))
traffic_A = float(input('Scenario A - Traffic Increment (µg/m3): '))
time_A = float(input('Scenario A - Time near road (minutes): '))
breathing_A = float(input('Scenario A - Breathing rate (m3/min): '))

# Scenario B Inputs
background_B = float(input('Scenario B - Background PM2.5 (µg/m3): '))
traffic_B = float(input('Scenario B - Traffic Increment (µg/m3): '))
time_B = float(input('Scenario B - Time near road (minutes): '))
breathing_B = float(input('Scenario B - Breathing rate (m3/min): '))
```

```python
# Calculation Function
def calculate_dose(background, traffic, time, breathing):
    C_total = background + traffic
    dose = C_total * breathing * time
    return C_total, dose

C_A, dose_A = calculate_dose(background_A, traffic_A, time_A, breathing_A)
C_B, dose_B = calculate_dose(background_B, traffic_B, time_B, breathing_B)

if dose_A != 0:
    reduction_percent = ((dose_A - dose_B) / dose_A) * 100
else:
    reduction_percent = 0

print('\n--- Results ---')
print('Scenario A - Total Concentration:', C_A, 'µg/m3')
print('Scenario A - Inhaled Dose:', dose_A, 'µg')

print('\nScenario B - Total Concentration:', C_B, 'µg/m3')
print('Scenario B - Inhaled Dose:', dose_B, 'µg')

print('\nDose Reduction from A to B:', reduction_percent, '%')
```
